```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure `df` is already available and preprocessed
# Assuming 'Date' is a datetime column and sorted
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by=['Company Name', 'Date'], inplace=True)

# Define quantiles and corresponding colors
quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
quantile_colors = ['blue', 'green', 'red', 'purple', 'orange']  # Assign a color to each quantile

# Compute quantiles
quantile_results = df.groupby('Company Name')['Close'].quantile(quantiles).unstack()

# Plot for each company
for company, company_data in df.groupby('Company Name'):
    plt.figure(figsize=(12, 6))

    # Plot the time series of closing prices
    plt.plot(company_data['Date'], company_data['Close'], label=f'{company} Close Prices', alpha=0.6, color='black')

    # Plot quantile lines with distinct colors
    for quantile, color in zip(quantiles, quantile_colors):
        quantile_value = quantile_results.loc[company, quantile]
        plt.axhline(y=quantile_value, linestyle='--', alpha=0.8, color=color, label=f'{int(quantile * 100)}th Percentile')

    plt.title(f'Empirical Quantiles for {company} Close Prices')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid()
    plt.show()
```

# KDE to estimate CDF and PDF

```python
from scipy.stats import gaussian_kde

for company, company_data in df.groupby('Company Name'):
    # Calculate log returns
    log_returns = np.log(company_data['Close'] / company_data['Close'].shift(1)).dropna()
    
    # Perform KDE
    kde = gaussian_kde(log_returns)
    
    # Generate values for PDF and CDF
    x = np.linspace(log_returns.min(), log_returns.max(), 1000)
    pdf = kde.evaluate(x)
    cdf = np.cumsum(pdf) / np.sum(pdf)  # Approximate CDF from PDF

    # Plot PDF
    plt.figure(figsize=(12, 6))
    plt.plot(x, pdf, label='PDF (KDE)', color='blue')
    plt.hist(log_returns, bins=50, density=True, alpha=0.6, color='gray', label='Empirical Data')
    plt.title(f'KDE PDF for {company} Log Returns')
    plt.xlabel('Log Returns')
    plt.ylabel('Density')
    plt.legend()
    plt.grid()
    plt.show()

    # Plot CDF
    plt.figure(figsize=(12, 6))
    plt.plot(x, cdf, label='CDF (KDE)', color='red')
    plt.title(f'KDE CDF for {company} Log Returns')
    plt.xlabel('Log Returns')
    plt.ylabel('Cumulative Probability')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import numpy as np
from scipy.stats import norm

def variance_ratio_test(series, lag=2):
    """
    Perform the Variance Ratio Test to check if a series follows a random walk.
    
    Args:
        series (array-like): The log return series.
        lag (int): The lag parameter for aggregation.

    Returns:
        tuple: Test statistic and p-value.
    """
    n = len(series)
    mu = np.mean(series)

    # Compute the numerator (variance of lagged differences)
    lagged_differences = series[lag:] - series[:-lag]
    var_lagged = np.var(lagged_differences, ddof=1)

    # Compute the denominator (variance of individual returns)
    var_individual = np.var(series, ddof=1)

    # Variance ratio
    vr = var_lagged / (lag * var_individual)

    # Compute test statistic
    z_stat = (vr - 1) / np.sqrt(2 * (2 * lag - 1) / (3 * lag * n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    return z_stat, p_value

# Example usage with log returns
for company, company_data in df.groupby('Company Name'):
    log_returns = company_data['Log Return'].dropna().values
    z_stat, p_value = variance_ratio_test(log_returns, lag=2)
    print(f"{company}: Variance Ratio Test Statistic={z_stat:.4f}, p-value={p_value:.4e}")
    if p_value < 0.05:
        print("  -> Reject null hypothesis: Evidence of time-varying volatility.")
    else:
        print("  -> Fail to reject null hypothesis: Volatility may be constant.")
```

```python
from scipy.stats import ks_2samp, norm, lognorm

for company, company_data in df.groupby('Company Name'):
    # Drop NaN values
    log_returns = company_data['Log Return'].dropna()

    # Fit Normal Distribution
    mu_norm, sigma_norm = norm.fit(log_returns)

    # Generate theoretical normal distribution
    ks_stat_norm, p_value_norm = ks_2samp(
        log_returns, 
        norm.rvs(loc=mu_norm, scale=sigma_norm, size=len(log_returns))
    )

    # Filter positive returns for log-normal
    positive_returns = log_returns[log_returns > 0]
    if len(positive_returns) > 0:
        shape_lognorm, loc_lognorm, scale_lognorm = lognorm.fit(positive_returns, floc=0)
        ks_stat_lognorm, p_value_lognorm = ks_2samp(
            positive_returns, 
            lognorm.rvs(shape_lognorm, loc_lognorm, scale_lognorm, size=len(positive_returns))
        )
    else:
        ks_stat_lognorm, p_value_lognorm = None, None

    # Print Results
    print(f'{company}: Kolmogorov-Smirnov Test Results')
    print(f'  Normal Distribution: KS Statistic={ks_stat_norm:.4f}, p-value={p_value_norm:.4e}')
    if p_value_norm < 0.05:
        print("    -> Reject null hypothesis: Returns do not follow normal distribution.")
    else:
        print("    -> Fail to reject null hypothesis: Returns may follow normal distribution.")

    if ks_stat_lognorm is not None:
        print(f'  Log-Normal Distribution: KS Statistic={ks_stat_lognorm:.4f}, p-value={p_value_lognorm:.4e}')
        if p_value_lognorm < 0.05:
            print("    -> Reject null hypothesis: Returns do not follow log-normal distribution.")
        else:
            print("    -> Fail to reject null hypothesis: Returns may follow log-normal distribution.")
```

```python
from scipy.stats import anderson

for company, company_data in df.groupby('Company Name'):
    # Drop NaN values
    log_returns = company_data['Log Return'].dropna()

    # Normal Distribution
    mu_norm, sigma_norm = norm.fit(log_returns)
    standardized_returns = (log_returns - mu_norm) / sigma_norm
    ad_stat_norm = anderson(standardized_returns, dist='norm')

    # Print Results
    print(f'{company}: Anderson-Darling Test Results')
    print(f'  Normal Distribution: Statistic={ad_stat_norm.statistic:.4f}')
    print(f'  Critical Values: {ad_stat_norm.critical_values}')
    print(f'  Significance Levels: {ad_stat_norm.significance_level}')

    if ad_stat_norm.statistic > ad_stat_norm.critical_values[-1]:
        print("    -> Reject null hypothesis: Returns do not follow normal distribution.")
    else:
        print("    -> Fail to reject null hypothesis: Returns may follow normal distribution.")

    # Log-Normal Distribution
    positive_returns = log_returns[log_returns > 0]
    if len(positive_returns) > 0:
        shape_lognorm, loc_lognorm, scale_lognorm = lognorm.fit(positive_returns, floc=0)
        standardized_positive_returns = (positive_returns - loc_lognorm) / scale_lognorm
        ad_stat_lognorm = anderson(standardized_positive_returns, dist='logistic')

        print(f'  Log-Normal Distribution: Statistic={ad_stat_lognorm.statistic:.4f}')
        print(f'  Critical Values: {ad_stat_lognorm.critical_values}')
        print(f'  Significance Levels: {ad_stat_lognorm.significance_level}')
        if ad_stat_lognorm.statistic > ad_stat_lognorm.critical_values[-1]:
            print("    -> Reject null hypothesis: Returns do not follow log-normal distribution.")
        else:
            print("    -> Fail to reject null hypothesis: Returns may follow log-normal distribution.")
```

```python
# Calculate lagged log returns using transform to maintain index alignment
df['Lagged Log Return'] = df.groupby('Company Name')['Log Return'].transform(lambda x: x.shift(1))
```

```python
# Drop rows with missing values in 'Log Return' or 'Lagged Log Return'
df.dropna(subset=['Log Return', 'Lagged Log Return'], inplace=True)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import TheilSenRegressor
import statsmodels.api as sm
import numpy as np
import pandas as pd

# Prepare a results dataframe to store model outputs
results = []

# Get unique company names
companies = df['Company Name'].unique()

for company in companies:
    # Subset data for the company
    company_data = df[df['Company Name'] == company]
    
    # Define dependent and independent variables
    X = company_data['Lagged Log Return'].values.reshape(-1, 1)
    y = company_data['Log Return'].values
    
    # Remove NaN values
    mask = ~np.isnan(X).flatten() & ~np.isnan(y)
    X = X[mask]
    y = y[mask]
    
    # OLS Regression
    X_with_const = sm.add_constant(X)  # Add constant term for OLS
    ols_model = sm.OLS(y, X_with_const).fit()
    ols_pred = ols_model.predict(X_with_const)
    
    # Extract OLS stats
    ols_intercept = ols_model.params[0]
    ols_slope = ols_model.params[1]
    ols_r_squared = ols_model.rsquared
    ols_p_value = ols_model.pvalues[1]
    
    # Theil-Sen Regression
    theil_sen_model = TheilSenRegressor().fit(X, y)
    theil_sen_pred = theil_sen_model.predict(X)
    
    # Extract Theil-Sen stats
    theil_sen_intercept = theil_sen_model.intercept_
    theil_sen_slope = theil_sen_model.coef_[0]
    
    # Append results to the list
    results.append({
        'Company Name': company,
        'OLS Intercept': ols_intercept,
        'OLS Slope': ols_slope,
        'OLS R-squared': ols_r_squared,
        'OLS P-value': ols_p_value,
        'Theil-Sen Intercept': theil_sen_intercept,
        'Theil-Sen Slope': theil_sen_slope
    })
    
    # Plotting the data and regression lines
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data Points', alpha=0.6)
    plt.plot(X, ols_pred, color='red', label='OLS Regression Line', linewidth=2)
    plt.plot(X, theil_sen_pred, color='green', label='Theil-Sen Regression Line', linewidth=2)
    plt.title(f'Regression Comparison for {company}', fontsize=14)
    plt.xlabel('Lagged Log Return', fontsize=12)
    plt.ylabel('Log Return', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(alpha=0.3)
    plt.show()

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

```
